```python
import os
import numpy as np
import cv2
from PIL import Image, ImageDraw, ImageEnhance, ImageFilter, ImageFont, ExifTags
import json
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm.notebook import tqdm
```

```python
#resize images in a folder to a specific size
print(os.getcwd())
ycb_models_path = "models/ycb"
resized_ycb_models_path = "models/ycb_resized"
if not os.path.exists(resized_ycb_models_path):
    os.makedirs(resized_ycb_models_path)
# in each folder in ycb_models_path, resize all images to 720 width
# get only final folder name
object_names = [f.name for f in os.scandir(ycb_models_path) if f.is_dir()]
object_paths = [f.path for f in os.scandir(ycb_models_path) if f.is_dir()]
object_names = sorted(object_names)
object_paths = sorted(object_paths)
print(object_names)
print(object_paths)

# jump objects, so as to not resize the whole dataset
object_jump = 3

for i in range(len(object_names)):
    object_name = object_names[i]
    object_path = object_paths[i]
    if not os.path.exists(os.path.join(resized_ycb_models_path, object_name)):
        os.makedirs(os.path.join(resized_ycb_models_path, object_name))
    
    filenames = os.listdir(object_path)
    filenames = sorted(filenames)
    
    print(filenames)
    
    object_number = len(range(0, len(filenames), object_jump))
    progress_bar = tqdm(total=object_number, desc=f"Resizing {object_name} in {object_path}")
    
    for j in range(0, len(filenames), object_jump):
        filename = filenames[j]
        if not (filename.endswith(".jpg") or filename.endswith(".png")):
            continue
        img = Image.open(os.path.join(object_path, filename))
        # Correct orientation
        for orientation in ExifTags.TAGS.keys() : 
            if ExifTags.TAGS[orientation]=='Orientation' : break 
        
        exif=dict(img._getexif().items())

        if   exif[orientation] == 3 : 
            img=img.rotate(180, expand=True)
        elif exif[orientation] == 6 : 
            img=img.rotate(270, expand=True)
        elif exif[orientation] == 8 : 
            img=img.rotate(90, expand=True)
        
        # Resize
        new_width= 720
        aspect_ratio = img.height / img.width
        new_height = int(new_width * aspect_ratio)
        img = img.resize((new_width, new_height))
        
        save_path = os.path.join(resized_ycb_models_path, object_name, filename)
        img.save(save_path)
        progress_bar.update(1)
    progress_bar.close()
```

### Resize PBM masks ###

```python
for i in range(len(object_names)):
    object_name = object_names[i]
    object_path = object_paths[i]
    mask_path = os.path.join(object_path, "masks")
    try:
        mask_files = os.listdir(mask_path)
    except:
        print(f"No masks found for {object_name}")
        continue
    
    # delete _mask.pmb to order it correctly
    mask_files = [mask_file[:-9] for mask_file in mask_files]
    mask_files = sorted(mask_files)
    
    # return the original mask path
    mask_files = [mask_file + "_mask.pbm" for mask_file in mask_files]
    print(mask_files)
    
    resized_mask_path = os.path.join(resized_ycb_models_path, object_name, "masks")
    if not os.path.exists(os.path.join(resized_ycb_models_path, object_name, "masks")):  
        os.makedirs(os.path.join(resized_ycb_models_path, object_name, "masks"))
    
    object_number = len(os.listdir(mask_path)) // object_jump
    progress_bar = tqdm(total=object_number, desc=f"Resizing masks from {object_name} in {mask_path}")
    
    for j in range(0, len(os.listdir(mask_path)), object_jump):
        
        
        
        filename = mask_files[j]
        if not (filename.endswith(".pbm")):
            continue
        img = Image.open(os.path.join(mask_path, filename))
        # Correct orientation
        try:
            for orientation in ExifTags.TAGS.keys() : 
                if ExifTags.TAGS[orientation]=='Orientation' : break 
            
            exif=dict(img._getexif().items())

            if   exif[orientation] == 3 : 
                img=img.rotate(180, expand=True)
            elif exif[orientation] == 6 : 
                img=img.rotate(270, expand=True)
            elif exif[orientation] == 8 : 
                img=img.rotate(90, expand=True)
        except:
            pass
        
        # Resize
        new_width= 720
        aspect_ratio = img.height / img.width
        new_height = int(new_width * aspect_ratio)
        img = img.resize((new_width, new_height))
        img.save(os.path.join(resized_mask_path, filename))
        
        progress_bar.update(1)
        
    progress_bar.close()
```
